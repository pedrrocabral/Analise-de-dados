```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

QUESTION: O tempo de Trabalho interfere na Renda?

```python
data = pd.read_csv(r'C:\Cursos\Analise_de_dados\Cases\Recursos Humanos\Salary_Data.csv')
```

```python
data = data.rename(columns={
    'YearsExperience': 'Anos de Experiência',
    'Salary': 'Salário'
})
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Análise da Renda', size=15)
sns.kdeplot( data['Salário'])
plt.show()
```

```python
plt.figure( figsize=(10, 6) )
plt.title('Análise dos Anos de Experiência', size=15)
sns.kdeplot( data['Anos de Experiência'])
plt.show()
```

```python
# Distribuição dos Salários
plt.figure( figsize=(10, 5) )
sns.boxplot( data['Salário'] )
plt.show()
```

```python
# Distribuição dos Anos de Experiência
plt.figure( figsize=(10, 5) )
sns.boxplot( data['Anos de Experiência'] )
plt.show()
```

```python
plt.figure( figsize=(10, 5) )
sns.regplot(data=data, x='Salário', y='Anos de Experiência')
plt.show()
```

ANSWER: Olhando os dados acima, o tempo de experiência das pessoas está ligado diretamente ao quanto elas vão receber
        quanto mais tempo de experiência maior o seu salário.
