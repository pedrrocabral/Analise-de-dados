```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(r'C:\Cursos\Analise_de_dados\Cases\Educação\StudentsPerformance+.csv')
```

```python
# Distribuição dos Participantes Por 'gender'
analise = data['gender'].value_counts(normalize=True) * 100
plt.figure(figsize=(10, 5))
plt.title('Distribuição dos Participantes Por Gênero', size=25)
plt.pie(
    analise.head(),
    labels=analise.index[:2],
    shadow=True,
    startangle=-50,
    explode=(0.04, 0.04),
    colors=['#F781F3', '#819FF7'],
    autopct='%.1f%%',
    textprops={'color': 'black', 'size': 20})
plt.show()
```

```python
# Boxplot de 'math score' por 'gender'
sns.boxplot(data=data, x='math score', y='gender');
```

```python
data.groupby(by='gender').describe()['math score'].reset_index()
```

```python
# Boxplot de 'reading score' por 'gender'
sns.boxplot(data=data, x='reading score', y='gender');
```

```python
data.groupby(by='gender').describe()['writing score'].reset_index()
```

```python
# Boxplot de 'reading score' por 'gender'
sns.boxplot(data=data, x='reading score', y='gender');
```

```python
data.groupby(by='gender').describe()['reading score'].reset_index()
```
